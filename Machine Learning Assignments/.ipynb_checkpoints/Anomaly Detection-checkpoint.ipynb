```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.covariance import EllipticEnvelope
from sklearn.svm import SVC
```

### Anomaly Detection

```python
d=loadmat("C://Users/Saba Naseem/Desktop/Machine Learning/Datasets/ex8data1.mat")
d.keys()
```

```python
x= d['X']
x.shape
```

```python
plt.figure(figsize=[7,5])
plt.scatter(x[:,0], x[:,1], marker='x')
plt.title("Outlier detection")
plt.xlabel('Latency (ms)')
plt.ylabel('Throughput (mb/s)')
plt.show()
```

Fitting data in model.

```python
ad = EllipticEnvelope()
ad.fit(x)
```

```python
predict = ad.decision_function(x).ravel()
predict.shape
```

Selecting the threshold

```python
percentile = 1.9
threshold = np.percentile(predict, percentile)
outliers = predict < threshold
```

```python
plt.figure(figsize=[10,7])

xx, yy = np.meshgrid(np.linspace(0, 30, 200), np.linspace(0, 30, 200))
Z = ad.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.scatter(x[:,0], x[:,1], marker='x')

plt.scatter(x[outliers][:,0], x[outliers][:,1], c='r', marker='x', linewidths=2)

plt.contour(xx, yy, Z, levels=[threshold], linewidths=2, colors='black', linestyles='dotted')

percent=[0.5,0.7,0.9,1.1,1.3,1.5,1.7]
for val in percent:
    
    threshold = np.percentile(predict, val)
    plt.contour(xx, yy, Z, levels=[threshold], linewidths=2, colors='red', linestyles='dotted')

plt.title("Outlier detection")
plt.xlabel('Latency (ms)')
plt.ylabel('Throughput (mb/s)')
plt.show()
```
