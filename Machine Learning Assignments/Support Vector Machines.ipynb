```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.optimize import minimize
from sklearn.svm import SVC
```

**Example Dataset 1.**

```python
d=loadmat("C://Users/Saba Naseem/Desktop/Machine Learning/Datasets/ex6data1.mat")
```

```python
x, y =d['X'], d['y']

d['X'].shape, d['y'].shape
```

Let's plot dataset.

```python
pos= (y==1).ravel()
neg= (y==0).ravel()

plt.figure(figsize=[7,5])
plt.scatter(x[pos,0], x[pos,1], marker='+',c='b', s=60) # c=color, s=size
plt.scatter(x[neg,0], x[neg,1], marker='o',c='y',s=60)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

### SVM with Linear Kernels

SVM with Linear Kernels means 'No Kernel' or 'Logistic Regression'

```python
# SVC=Support Vector Classification.

svm=SVC(C=1.0, kernel='linear')
svm.fit(x,y.ravel())
```

```python
vec=svm.support_vectors_
```

```python
plt.figure(figsize=[7,5])
plt.scatter(x[pos,0], x[pos,1], marker='+',c='b', s=60) # c=color, s=size
plt.scatter(x[neg,0], x[neg,1], marker='o',c='y',s=60)

h=0.02
pad=0.25
x_min, x_max = x[:, 0].min()-pad, x[:, 0].max()+pad
y_min, y_max = x[:, 1].min()-pad, x[:, 1].max()+pad
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.2)

plt.scatter(vec[:,0], vec[:,1], c='k', marker='|', s=100, linewidths='1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title("SVM Decision Boundary with C = 1")
plt.show()
```
